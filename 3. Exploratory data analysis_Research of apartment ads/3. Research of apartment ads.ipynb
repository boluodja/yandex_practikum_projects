## Добавление вспомогательных фич

__Добавление цены за метр квадратный__

```python
data['ppsm'] = round(data['last_price'] / data['total_area'] / 1000 , 0)
```

__Добавление дня недели, месяца, года публикации__

```python
#изменение формата даты
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
#добавление столбца
data['year_exposition'] = data['first_day_exposition'].dt.year 
data['month_exposition'] = data['first_day_exposition'].dt.month 
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
#вывод 
data.head()
```

__Добавление категории этажа квартиры__

```python
# Функция присвоения категории этажа
def floor_definition(row):
    if row['floor'] == row['floors_total']:
        return 'последний'
    elif row['floor'] == 1:
        return 'первый'   
    else:
        return 'другой'
    
# Функция присвоения id этажа для графиков
def id_floor(row):
    id = {'первый' : 1,
           'другой' : 2,
           'последний': 3}
    return id.get(row['floor_definition'])
```

```python
data['floor_definition'] = data.apply(floor_definition, axis = 1)
data['id_floor'] = data.apply(id_floor, axis = 1) 
data.head()
```

__Добавление соотношений площади жилой к общей, а также кухни к общей__

```python
data['kitchen_to_total_area'] = data['kitchen_area'] / data['total_area']
data['living_to_total_area'] = data['living_area'] / data['total_area']
data.head()
```

## Изучение данных:

__Функция для вычищения данных от выбросов__

```python
def delimitation(row, left_limit = False, right_limit = False):
    # определим первый и третий квантиль
    q1 = np.nanquantile(row, 0.25)
    q3 = np.nanquantile(row, 0.75)
    # определим межквантильный размах 
    iqr = q3 - q1
    
    # установим ограничения с учетом того, что исследуемые параметры не могут принимать отрицательные значения
    if left_limit:
        left_bound = max(q1 - 1.5 * iqr, 0)
    else:
        left_bound = 0
    if right_limit:
        right_bound = min(q3 + 1.5 * iqr, np.nanquantile(row, 0.95))
    else:
        right_bound = row.max()
    return left_bound, right_bound
```



### Площадь

Очистим данные от выбросов.

```python
left_bound, right_bound = delimitation(data['total_area'],left_limit = True, right_limit = True)
f'Интервал, вычищенный от выбросов: {round(left_bound)} - {round(right_bound)}'
```

```python
good_data = data.query('@left_bound <= total_area <= @right_bound')
good_data.hist('total_area', bins = 'auto', color = 'black')
```

__Вспомогательные функции__

Создадим функцию для построения матрицы коэффициентов корелляции Пирсона

```python
def pirson_matrix(filtered_data):
    # зададим размер
    plt.figure(figsize=(5,5))
    # cоздадим массив из коэффициентов корелляций Пирсона и отрежем данные ниже диагонали
    mask = np.tril(filtered_data.corr())
    # построим график
    sns.heatmap(filtered_data.corr(),
                cmap='coolwarm', center=0, annot=True, mask = mask, cbar = False, square = True, fmt = '.2f')
    plt.title('Матрица коэффифентов корелляции Пирсона', fontsize=14)
    return plt.show()
```

Создадим функцию для построения диаграмм рассеивания

```python
''' Параметры:
shedules_titles - массив из названий графиков
shedules_count - количество графиков
x1 - список из названий столбцов, которые будем строить по оси абссисс
y1 - список из названий столбцов, которые будем строить по оси ординат
df - исследуемый датафрэйм
with_median - принимает булево значение в зависимотси от необходимости добавления 
            медианных значений для дискретных переменных (например медианы стоимостей квартир 
            определенных этажах)
nonparametric - задаем True, если необходима непараметрическая регрессия 
hue_location - нужна для построения графика с выделением конкретной локации, например, центра Санкт-Петербурга
'''       
        
def graphic_builder(shedules_titles, shedules_count, x1, y1, df, with_median, nonparametric, hue_location = False):
    
    # функция нужна для выделения из массива данных центра Санкт-Петербурга в отдельный df (для пункта 7)
    if hue_location == True:
        df_specific_location = df[df['spb_center'] == True]
        df = df[df['spb_center'] == False]
        
    fig = plt.figure()
    fig.set_figwidth(14)
    fig.set_figheight(14)

        
# строим необходимое количество графиков (макс - 4 шт)
    for i in range(shedules_count):     
        ax1 = fig.add_subplot(2, 2, i + 1)
        sns.regplot(x = x[i], y = y[i], data = df,
                    lowess = nonparametric[i], 
                    scatter_kws = {"color": "LightGray", "alpha":0.01}, line_kws={"color": "black"},
                    ax = ax1)
        
# на случай, если надо для дискретных значений х добавить по оси ординат медианы
        if with_median[i]:
             sns.regplot(x = x[i], y = y[i], data = df, 
                        x_estimator = np.mean,
                        color = 'black',
                        ax = ax1)
        ax1.set_title(shedules_titles[i], fontsize=12)
        

# строим необходимые графики по центру Санкт-Петербурга, по сути это повтор функции выше только по центру СПБ
        if hue_location == True: 
                sns.regplot(x = x[i], y = y[i], data = df_specific_location,
                            lowess = nonparametric[i], 
                            scatter_kws = {"color": "Aqua", "alpha":0.01}, line_kws={"color": "DarkGreen"},
                            ax = ax1)
                if with_median[i]:
                    sns.regplot(x = x[i], y = y[i], data = df_specific_location, 
                                x_estimator = np.mean,
                                color = 'DarkGreen',
                                ax = ax1)    
    return plt.show()  
```
